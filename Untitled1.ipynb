```python
# ===========================================
# APLIKASI MINI DATA MAHASISWA (MENU INTERAKTIF)
# DENI HARTANTO KURNIAWAN IE.24.C5
# ===========================================

def input_data():
    n = int(input("Jumlah mahasiswa: "))
    data = []

    for i in range(n):
        print(f"\nData mahasiswa ke-{i+1}")
        nama = input("Nama: ")
        nilai = float(input("Nilai: "))
        data.append((nama, nilai))

    print("\nData berhasil dimasukkan!")
    return data


def hitung_rata(data):
    if not data:
        return 0
    return sum([x[1] for x in data]) / len(data)


def nilai_tertinggi(data):
    if not data:
        return None
    return max(data, key=lambda x: x[1])


def kategori(nilai):
    if nilai >= 85:
        return "A"
    elif nilai >= 70:
        return "B"
    elif nilai >= 55:
        return "C"
    else:
        return "D"


def urutkan_data(data):
    return sorted(data, key=lambda x: x[1], reverse=True)


def bawah_rata(data, rata):
    return [m for m in data if m[1] < rata]


def buat_laporan(data):
    if not data:
        return "Belum ada data mahasiswa."

    rata = hitung_rata(data)
    tertinggi = nilai_tertinggi(data)
    data_urut = urutkan_data(data)
    bawah = bawah_rata(data, rata)

    report = "===== LAPORAN DATA MAHASISWA =====\n\n"
    report += "Data Mahasiswa:\n"
    for nama, nilai in data:
        report += f"- {nama:15} | {nilai:5} | Kategori: {kategori(nilai)}\n"

    report += f"\nRata-rata nilai: {rata:.2f}\n"
    report += f"Nilai tertinggi: {tertinggi[0]} ({tertinggi[1]})\n\n"

    report += "Data urut berdasarkan nilai:\n"
    for nama, nilai in data_urut:
        report += f"- {nama:15} : {nilai}\n"

    report += "\nMahasiswa di bawah rata-rata:\n"
    if bawah:
        for nama, nilai in bawah:
            report += f"- {nama:15} : {nilai}\n"
    else:
        report += "(Tidak ada)\n"

    return report


def simpan_laporan(filename, report):
    with open(filename, "w") as f:
        f.write(report)
    print(f"\nâœ” Laporan berhasil disimpan ke file: {filename}")


# =========================
# MENU INTERAKTIF
# =========================

def main():
    data = []
    while True:
        print("\n=========== MENU APLIKASI ===========")
        print("1. Input data mahasiswa")
        print("2. Tampilkan laporan")
        print("3. Tampilkan data urut nilai")
        print("4. Tampilkan mahasiswa di bawah rata-rata")
        print("5. Simpan laporan ke file")
        print("0. Keluar")

        pilihan = input("Pilih menu: ")

        if pilihan == "1":
            data = input_data()

        elif pilihan == "2":
            print("\n===== LAPORAN =====")
            print(buat_laporan(data))

        elif pilihan == "3":
            print("\n===== DATA URUT BERDASARKAN NILAI =====")
            for nama, nilai in urutkan_data(data):
                print(f"- {nama:15} : {nilai}")

        elif pilihan == "4":
            rata = hitung_rata(data)
            bawah = bawah_rata(data, rata)
            print("\n===== NILAI DI BAWAH RATA-RATA =====")
            if bawah:
                for nama, nilai in bawah:
                    print(f"- {nama:15} : {nilai}")
            else:
                print("(Tidak ada data)")

        elif pilihan == "5":
            report = buat_laporan(data)
            simpan_laporan("laporan_mahasiswa.txt", report)

        elif pilihan == "0":
            print("\nTerima kasih! Aplikasi selesai.")
            break

        else:
            print("Pilihan tidak valid, coba lagi!")

# Jalankan program
if __name__ == "__main__":
    main()
```
